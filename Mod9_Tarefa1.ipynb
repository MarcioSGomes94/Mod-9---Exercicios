### 1. Idade media das mães e dos pais por município (coluna munResNome)

```python
resultados = sinasc_1.groupby('munResNome')[['IDADEMAE', 'IDADEPAI', 'QTDFILVIVO']].mean()

resultados_sorted = resultados.sort_values(by='QTDFILVIVO', ascending=False)

resultados_sorted
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python

sinasc_raw['DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])

nascimentos_por_municipio = sinasc_raw['munResNome'].value_counts()

municipio_menos_nascimentos = nascimentos_por_municipio.idxmin()
nascimentos_municipio_menos = nascimentos_por_municipio.min()

print(f'O município com menos nascimentos em 2019 foi {municipio_menos_nascimentos} com {nascimentos_municipio_menos} nascimentos.')
```

```python
municipio_stats = sinasc_raw.groupby('munResNome')[['IDADEMAE', 'IDADEPAI']].agg(
    idade_media_mae=('IDADEMAE', 'mean'),
    idade_maxima_mae=('IDADEMAE', 'max'),
    idade_minima_mae=('IDADEMAE', 'min'),
    idade_media_pai=('IDADEPAI', 'mean'),
    idade_maxima_pai=('IDADEPAI', 'max'),
    idade_minima_pai=('IDADEPAI', 'min')
)
municipio_stats.loc[municipio_menos_nascimentos].to_frame()
```
